# Functions

```python
def read_data(path, name, pct = False, z = False):
    test = pd.read_excel(path)
    test['Time'] = pd.DatetimeIndex(pd.to_datetime(test['Time'])).to_period("Y")
    
    test_melt = test.groupby('Time').last().reset_index().melt(id_vars=["Time"], var_name="Company", value_name=name)
        
    test_melt[name+' Change'] = test_melt.groupby(['Company'])[name].transform(lambda x: x.diff())
    
    if pct:
        test_melt[name+' Change Pct'] = test_melt.groupby(['Company'])[name].transform(lambda x: x.pct_change(fill_method=None))
        
    if z:
        test_melt['Z-norm: '+name] = test_melt.groupby(['Time'])[name].transform(lambda x: zscore(x, nan_policy='omit'))
        test_melt['Z-norm: '+name+' Change'] = test_melt.groupby(['Time'])[name+' Change'].transform(lambda x: zscore(x, nan_policy='omit'))
        
    return test_melt

def read_data_yearly(path, name, pct= False, z= False):
    test = pd.read_excel(path)
    test_melt = test.melt(id_vars=["Time"], var_name="Company", value_name=name)
    
    test_melt[name+' Change'] = test_melt.groupby(['Company'])[name].transform(lambda x: x.diff())
    
    test_melt['Time'] = pd.DatetimeIndex(pd.to_datetime(test_melt['Time'].astype(str))).to_period("Y")
    
    if pct:
        test_melt[name+' Change Pct'] = test_melt.groupby(['Company'])[name].transform(lambda x: x.pct_change(fill_method=None))
    
    if z:
        test_melt['Z-norm: '+name] = test_melt.groupby(['Time'])[name].transform(lambda x: zscore(x, nan_policy='omit'))
        test_melt['Z-norm: '+name+' Change'] = test_melt.groupby(['Time'])[name+' Change'].transform(lambda x: zscore(x, nan_policy='omit'))
        
    return test_melt
```

### Compute Vola

```python
def ssd(data, bw, sc, name):
    data[name] = float('nan')
    for company in np.unique(data['Company']):
        for year in np.unique(data[data['Company'] == company]['Time']):
            ind = data[(data['Company'] == company) & (data['Time'] == year)].index
            if data.loc[(data['Company'] == company) & (data['Time'] <= year) & (data['Time'] > year-bw), sc].isnull().values.any():
                #print("NaN in window")
                pass
            else:
                data.loc[ind, name] = data.loc[(data['Company'] == company) & (data['Time'] <= year) & (data['Time'] > year-bw), sc].std()
    return data

def our(data, bw, sc, name):
    data[name] = float('nan')
    for company in np.unique(data['Company']):
        for year in np.unique(data[data['Company'] == company]['Time']):
            ind = data[(data['Company'] == company) & (data['Time'] == year)].index
            if data.loc[(data['Company'] == company) & (data['Time'] <= year) & (data['Time'] >= year-bw), sc].isnull().values.any():
                #print("NaN in window")
                pass
            else:
                data.loc[ind, name] = np.sqrt((data.loc[(data['Company'] == company) & (data['Time'] <= year) & (data['Time'] >= year-bw), sc].diff()**2).mean())
    return data
```

## Drop Extreme Values
We use Z-Score for extreme value detection: arbitrary Threshold = 3.5

### scale absolute values (CapEx, Employees, Net Income, Net Sales, Total Assets) with Market Value

```python
def scale_size(data, name):
    data[name+"/MV"] = data[name]/data['Market Value']
    data[name+'/MV Change'] = data.groupby(['Company'])[name+'/MV'].transform(lambda x: x.diff())
    data[name+'/MV Change Pct'] = data.groupby(['Company'])[name+'/MV'].transform(lambda x: x.pct_change())
    data['Z-norm: '+name+'/MV'] = data.groupby(['Time'])[name+'/MV'].transform(lambda x: zscore(x, nan_policy='omit'))
    data['Z-norm: '+name+'/MV Change'] = data.groupby(['Time'])[name+'/MV Change'].transform(lambda x: zscore(x, nan_policy='omit'))
    return data
```

```python
data = scale_size(data,'CapEx')
data = scale_size(data,'Employees')
data = scale_size(data,'Net Income')
data = scale_size(data,'Net Sales')
data = scale_size(data,'Total Assets')
```
