# 1. Functions

```python
def portfolio(data, var, perc, year, date_end, prices_me):
    # split in lowest and highest and create benchmark set (here: all avalaible companies)
    low_vol = data[(data[var] < perc) & (data['Time'] == year)]
    high_vol = data[(data[var] >= 1-perc) & (data['Time'] == year)]
    bm = data[data['Time'] == year]
    

    # set of unique companies with available score of interest in the given year
    low_comp = pd.DataFrame(low_vol.groupby('Time')['Company'].apply(lambda x: list(np.unique(x))))
    high_comp = pd.DataFrame(high_vol.groupby('Time')['Company'].apply(lambda x: list(np.unique(x))))
    bm_comp = pd.DataFrame(bm.groupby('Time')['Company'].apply(lambda x: list(np.unique(x))))
    
    
    # add help columns for date comparison
    low_comp['month'] = low_comp.index.month
    low_comp['year'] = low_comp.index.year
    low_comp['date'] = pd.to_datetime(low_comp[['year','month']].assign(day=1)).dt.to_period('D')
    
    high_comp['month'] = high_comp.index.month
    high_comp['year'] = high_comp.index.year
    high_comp['date'] = pd.to_datetime(high_comp[['year','month']].assign(day=1)).dt.to_period('D')
    
    bm_comp['month'] = bm_comp.index.month
    bm_comp['year'] = bm_comp.index.year
    bm_comp['date'] = pd.to_datetime(bm_comp[['year','month']].assign(day=1)).dt.to_period('D')
    
    # set date of score comparison
    date2 = low_comp.iloc[0,3]
 
    # add help variables
    pf_low = pd.DataFrame()
    pf_low_clean = []
    
    pf_high = pd.DataFrame()
    pf_high_clean = []
    
    pf_bm = pd.DataFrame()
    pf_bm_clean = []
    
    ind = 0
    comp_help = []
    comp_help_high = []
    comp_help_bm = []
    
    # low: find companies that also have price data available throughout the time of interest
    for company in low_comp.iloc[0,0]:
        if prices_me[(prices_me['Time'] < date_end) & (prices_me['Time'] >= (date2 + 4*30))][company].isnull().values.any():
            pass
        else:
            comp_help.append(company)
    #pf_low_clean.append(comp_help)  
    
    # high: find companies that also have price data available throughout the time of interest
    for company in high_comp.iloc[0,0]:
        if prices_me[(prices_me['Time'] < date_end) & (prices_me['Time'] >= (date2 + 4*30))][company].isnull().values.any():
            pass
        else:
            comp_help_high.append(company)
    #pf_high_clean.append(comp_help_high)  
    
    # benchmark: find companies that also have price data available throughout the time of interest
    for company in bm_comp.iloc[0,0]:
        if prices_me[(prices_me['Time'] < date_end) & (prices_me['Time'] >= (date2 + 4*30))][company].isnull().values.any():
            pass
        else:
            comp_help_bm.append(company)
    #pf_bm_clean.append(comp_help_bm)  
    
    date = date2 + 4*30
    # for these subsets: compute portfolio value
    prices_low = prices_me[(prices_me['Time'] > date) & (prices_me['Time'] < (pd.to_datetime(date_end)).to_period('D'))][comp_help]
    prices_low = prices_low.set_index(prices_me[(prices_me['Time'] > date) & (prices_me['Time'] < (pd.to_datetime(date_end)).to_period('D'))]['Time'])
    pf_low['Portfolio Value'] = prices_low.sum(axis=1)
    
    prices_high = prices_me[(prices_me['Time'] > date) & (prices_me['Time'] < (pd.to_datetime(date_end)).to_period('D'))][comp_help_high]
    prices_high = prices_high.set_index(prices_me[(prices_me['Time'] > date) & (prices_me['Time'] < (pd.to_datetime(date_end)).to_period('D'))]['Time'])
    pf_high['Portfolio Value'] = prices_high.sum(axis=1)
    
    prices_bm = prices_me[(prices_me['Time'] > date) & (prices_me['Time'] < (pd.to_datetime(date_end)).to_period('D'))][comp_help_bm]
    prices_bm = prices_bm.set_index(prices_me[(prices_me['Time'] > date) & (prices_me['Time'] < (pd.to_datetime(date_end)).to_period('D'))]['Time'])
    pf_bm['Portfolio Value'] = prices_bm.sum(axis=1)
    
#    while date < (pd.to_datetime(date_end)).to_period('D'):
#        pf_value = 0
#        pf_value_high = 0
#        pf_value_bm = 0
#        
#        # low: add up value of companies as value-weighted portfolio value
#        for company in comp_help:
#            if np.isnan(prices_me[prices_me['Time'] == date][company].values):
#                print('WRONG: NaN as prices: --- Should not be reached')
#                print(company)
#            else:
#                pf_value += prices_me[prices_me['Time'] == date][company].values
#        
#        # high: add up value of companies as value-weighted portfolio value
#        for company in comp_help_high:
#            if np.isnan(prices_me[prices_me['Time'] == date][company].values):
#                print('WRONG: NaN as prices: --- Should not be reached')
#                print(company)
#            else:
#                pf_value_high += prices_me[prices_me['Time'] == date][company].values
#           
#        # benchmark: add up value of companies as value-weighted portfolio value
#        for company in comp_help_bm:
#            if np.isnan(prices_me[prices_me['Time'] == date][company].values):
#                print('WRONG: NaN as prices: --- Should not be reached')
#                print(company)
#            else:
#                pf_value_bm += prices_me[prices_me['Time'] == date][company].values
#                
#        #pf_low.loc[ind, 'Portfolio'] = year
#        pf_low.loc[ind, 'date'] = date
#        #pf_high.loc[ind, 'Portfolio'] = year
#        pf_high.loc[ind, 'date'] = date
#        #pf_bm.loc[ind, 'Portfolio'] = year
#        pf_bm.loc[ind, 'date'] = date
#        
#        try:
#            pf_low.loc[ind, 'Portfolio Value'] = pf_value
#            pf_high.loc[ind, 'Portfolio Value'] = pf_value_high
#            pf_bm.loc[ind, 'Portfolio Value'] = pf_value_bm
#        except:
#            # on weekends and bank holidays --> no prices data available
#            pf_low.loc[ind, 'Portfolio Value'] = float('nan')
#            pf_high.loc[ind, 'Portfolio Value'] = float('nan')
#            pf_bm.loc[ind, 'Portfolio Value'] = float('nan')
#        ind += 1
#        date+=1
    
    # low: drop weekends and bank holidays and assign return and portfolio value if started at 1
    pf_low = pf_low.dropna()
    pf_low['Portfolio_diff'] = pf_low.diff()['Portfolio Value']
    pf_low['return'] = 1 + pf_low['Portfolio_diff']/pf_low['Portfolio Value'].shift()
    pf_low['value'] = pf_low['return'].cumprod()
    #pf_dict_low = dict(zip(low_comp.index, pf_low_clean))
    
    # high: drop weekends and bank holidays and assign return and portfolio value if started at 1
    pf_high = pf_high.dropna()
    pf_high['Portfolio_diff'] = pf_high.diff()['Portfolio Value']
    pf_high['return'] = 1 + pf_high['Portfolio_diff']/pf_high['Portfolio Value'].shift()
    pf_high['value'] = pf_high['return'].cumprod()
    #pf_dict_high = dict(zip(high_comp.index, pf_high_clean))
    
    # bm: drop weekends and bank holidays and assign return and portfolio value if started at 1
    pf_bm = pf_bm.dropna()
    pf_bm['Portfolio_diff'] = pf_bm.diff()['Portfolio Value']
    pf_bm['return'] = 1 + pf_bm['Portfolio_diff']/pf_bm['Portfolio Value'].shift()
    pf_bm['value'] = pf_bm['return'].cumprod()
    #pf_dict_bm = dict(zip(bm_comp.index, pf_bm_clean))
    
    return pf_low.reset_index(), comp_help, pf_high.reset_index(), comp_help_high, pf_bm.reset_index(), comp_help_bm
```

```python
def viz(low, high, bm, comp):
    plt.figure(figsize=(15, 7.5))
    plt.plot(pd.to_datetime((low.Time).astype(str)), low.value, label = 'Low '+comp, color = 'orange')
    #plt.plot(low.index, low.value, label = 'Low '+comp, color = 'orange')
    plt.plot(pd.to_datetime((high.Time).astype(str)), high.value, label = 'High '+ comp, color = 'blue')
    plt.plot(pd.to_datetime((bm.Time).astype(str)), bm.value, label = 'Benchmark', color = 'black')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value')
    plt.title('Comparison of Low vs. High ' + comp)
    plt.legend()
    plt.show()
```
