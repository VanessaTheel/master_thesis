```python
def get_return(mv, comp, prices):
    try:
        # assign percentiles to market caps
        mv_comps = mv[comp]
        mv_rank = mv_comps.iloc[0,:].rank(pct=True)
    
        # put companies into tertiles
        tert1_comp = mv_rank[mv_rank < 1/3].index.values
        tert2_comp = mv_rank[(mv_rank >= 1/3) & (mv_rank < 2/3)].index.values
        tert3_comp = mv_rank[mv_rank >= 2/3].index.values
    
        tert1_weights = mv_rank[mv_rank < 1/3].values/mv_rank[mv_rank < 1/3].values.sum()
        tert2_weights = mv_rank[(mv_rank >= 1/3) & (mv_rank < 2/3)].values/mv_rank[(mv_rank >= 1/3) & (mv_rank < 2/3)].values.sum()
        tert3_weights = mv_rank[mv_rank >= 2/3].values/mv_rank[mv_rank >= 2/3].values.sum()
    
        help_ = prices[tert1_comp]
        help__ = prices[tert2_comp]
        help___ = prices[tert3_comp]
        
        #print(sum(tert1_weights))
        #print(sum(tert2_weights))
        #print(sum(tert3_weights))
        
    
        # calculate value weighted returns in each tertile
        ret1 = sum(help_.iloc[0,:].values*tert1_weights)
        ret2 = sum(help__.iloc[0,:].values*tert2_weights)
        ret3 = sum(help___.iloc[0,:].values*tert3_weights)
    
        return statistics.mean((ret1,ret2,ret3))
    except:
        return float('nan')
```

```python
def reg_ana(daten, mv, prices):
    low_comp = pd.DataFrame(daten.groupby('Time')['Company'].apply(lambda x: list(np.unique(x))))
    low_comp['month'] = low_comp.index.month
    low_comp['year'] = low_comp.index.year
    low_comp['date'] = pd.to_datetime(low_comp[['year','month']].assign(day=1)).dt.to_period('M')
    pf_low = pd.DataFrame()
    pf_low_clean = []
    ind = 0
    for year in low_comp.index:
        date2 = low_comp.loc[year,'date']
        
        #find subset of companies that have no NaN for month-end prices in the period of interest
        comp_help = []
        for company in low_comp.loc[year, 'Company']:
            if prices_me[(prices_me['date'] < date2 + 15) & (prices_me['date'] >= date2 + 3)][company].isnull().values.any():
                pass
            else:
                comp_help.append(company)
        #print(len(comp_help))
        pf_low_clean.append(comp_help)  
    
        # for this subset: compute portfolio value
        for i in range(3,16,1):
            date = date2 + i
            ret = get_return(mv[mv['date'] == (date-1)], comp_help, prices[prices['date'] == date])
            pf_low.loc[ind, 'Portfolio'] = year
            pf_low.loc[ind, 'date'] = date
            try:
                pf_low.loc[ind, 'return'] = ret
            except:
                pf_low.loc[ind, 'return'] = float('nan')
            ind += 1
            
    #pf_low['value'] = (1+pf_low['return']).cumprod()
    pf_dict_low = dict(zip(low_comp.index, pf_low_clean))
    
    pf_low = pf_low.dropna()
    
    #pf_low_reg = pf_low.merge(ff5, on = 'date', how = 'left')
    
    #y = pf_low_reg.loc[:,"return"]
    #x = sm.add_constant(pf_low_reg[exog_vars])
    #model = sm.OLS(y,x.astype(float), missing= 'drop')
    #model = model.fit()
    
    return pf_low, pf_dict_low
```

```python
#find set of companies for portfolio each year
def find_comp(daten, prices_me):
    low_comp = pd.DataFrame(daten.groupby('Time')['Company'].apply(lambda x: list(np.unique(x))))
    low_comp['month'] = low_comp.index.month
    low_comp['year'] = low_comp.index.year
    low_comp['date'] = pd.to_datetime(low_comp[['year','month']].assign(day=1)).dt.to_period('D')
    pf_low = pd.DataFrame()
    pf_low_clean = []
    ind = 0
    for year in low_comp.index:
        
        date2 = low_comp.loc[year,'date']
        
        #find subset of companies that have no NaN for month-end prices in the period of interest
        comp_help = []
        for company in low_comp.loc[year, 'Company']:
            if prices_me[(prices_me['Time'] < (date2 + (30+365))) & (prices_me['Time'] >= (date2 + 30))][company].isnull().values.any():
                pass
            else:
                comp_help.append(company)
        pf_low_clean.append(comp_help)
    return dict(zip(low_comp.index.astype(str), pf_low_clean))
```

```python
def viz(low, high, comp1, comp2):
    plt.figure(figsize=(10, 6))
    
    start = max(min(low.date), min(high.date))
    ff5_bm = ff5[ff5.date >= start]
    low_2 = low[low.date >= start]
    high_2 = high[high.date >= start]
    
    ff5_bm = ff5_bm.set_index('date')
    ff5_mkt = (1+ff5_bm['Mkt-RF']+ff5_bm['RF']).cumprod()
    
    low_2['value'] = (1+low_2['return']).cumprod()
    high_2['value'] = (1+high_2['return']).cumprod()
    
    plt.ylim(bottom=0, top=16)
    
    plt.plot(pd.to_datetime((low_2.date).astype(str)), low_2.value, label = comp1, color = '#005293')
    #plt.plot(low.index, low.value, label = 'Low '+comp, color = 'orange')
    plt.plot(pd.to_datetime((high_2.date).astype(str)), high_2.value, label = comp2, color = '#64a0c8')
    plt.plot(pd.to_datetime((ff5_mkt.index).astype(str)), ff5_mkt, label = 'Market', color = '#999999')
    
    #plt.plot(pd.to_datetime((bm.Time).astype(str)), bm.value, label = 'Benchmark', color = 'black')
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value')
    plt.title('Comparison of ' + comp1 + ' and ' + comp2)
    plt.legend()
    plt.show()
```

```python
def viz_fac(long, short, comp1, comp2):
    long_gg_short_bb = long['return'] - short['return']
    plt.figure(figsize=(15, 7.5))
    plt.plot(pd.to_datetime((long.date).astype(str)), (1+long_gg_short_bb).cumprod())
    plt.xlabel('Date')
    plt.ylabel('Portfolio Value')
    plt.title('Long = ' + comp1 + ' / Short = ' + comp2)
    plt.show()
```

## Check whether alpha - Fama French 5 Factor

```python
ff5
```

```python
exog_vars = ['Mkt-RF', 'SMB', 'HML', 'RMW', 'CMA']
```

```python
def abnormal_ret(data_long, ff, data_short = pd.DataFrame()):
    if not data_short.empty:
        data = pd.DataFrame(data_long['return'] - data_short['return'])
        data['date'] = data_long['date']
    else:
        data = data_long
    pf_low_reg = data.merge(ff, on = 'date', how = 'left')
    y = pf_low_reg.loc[:,"return"]- pf_low_reg.loc[:,"RF"]
    x = sm.add_constant(pf_low_reg[exog_vars])
    model = sm.OLS(y,x.astype(float), missing= 'drop')
    model = model.fit()
    print(model.summary())
    
    # compute alpha, mean return and vola and annualize
    alpha = model.params[0]
    alpha_a = (1+alpha)**12 -1
    mu_com_c = data['return'].mean()
    sigma_com_c = data['return'].std()
    mu_com_c_a = (1+mu_com_c)**12 -1
    sigma_com_c_a = np.sqrt(((1+mu_com_c)**2+sigma_com_c**2)**12-(1+mu_com_c)**24)
    
    # compute expected (monthly) sharpe ratio
    sr = y.mean()/y.std()
    
    # put together:
    metrics = [alpha, alpha_a, mu_com_c, mu_com_c_a, sigma_com_c, sigma_com_c_a, sr]
    
    return model, metrics
```
